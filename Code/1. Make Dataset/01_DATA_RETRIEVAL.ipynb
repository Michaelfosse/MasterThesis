```python
import pandas as pd

df = pd.read_csv("references/idaSearch_4_26_2024.csv")
df



```

```python
# Assuming df is your DataFrame

# Separate MRI and PET dataframes for comparison
mri_df = df[df['Modality'] == 'MRI']
pet_df = df[df['Modality'] == 'PET']

# Find subjects with matching ages in both MRI and PET datasets
# This creates a multi-index with levels ['Subject', 'Age'] and finds the intersection
common = mri_df.set_index(['Subject ID', 'Age']).index.intersection(pet_df.set_index(['Subject ID', 'Age']).index)

# Filter the original dataframe to only include rows that match the common subjects and ages
filtered_df_both_subjects = df.set_index(['Subject ID', 'Age']).loc[common].reset_index()

# Now, filtered_df contains only the rows where a subject has both an MRI and PET scan at the same age.

filtered_df_both_subjects['Subject ID'].nunique()
```

```python
filtered_df_pet = filtered_df_both_subjects[filtered_df_both_subjects['Modality'] == 'PET']
```

```python
# Filter to get rows where 'description' contains 'fdg', case-insensitively
filtered_df = filtered_df_pet[filtered_df_pet['Description'].str.contains('fdg', case=False, na=False)]

print(filtered_df)
```

```python
filtered_df['Subject ID'].nunique()
```

```python
# Calculate the frequency of each description
frequency = filtered_df['Description'].value_counts().rename('freq')

# Merge the frequency data back to the original DataFrame
filtered_df = filtered_df.merge(frequency.to_frame(), left_on='Description', right_index=True)

# Sort the DataFrame by frequency (descending order), then by description for tie-breaking
filtered_df = filtered_df.sort_values(by=['freq', 'Description'], ascending=[False, True])

# Optionally, drop the frequency column if it's no longer needed
filtered_df = filtered_df.drop(columns='freq')

print(filtered_df)
```

```python
# Sort the DataFrame first by 'Subject ID' and 'Type' where 'Processed' entries are prioritized
# We use a custom sort key that ranks 'Processed' highest
filtered_df['Type_sort'] = filtered_df['Type'].apply(lambda x: 0 if x == 'Processed' else 1)
filtered_df.sort_values(by=['Subject ID', 'Type_sort'], inplace=True)

# Drop the helper column after sorting
filtered_df.drop(columns=['Type_sort'], inplace=True)

# Now drop duplicates keeping the first occurrence, which will now prefer 'Processed'
filtered_df_unique = filtered_df.drop_duplicates(subset='Subject ID', keep='first')

print(filtered_df_unique)
```

```python
filtered_df_unique = filtered_df_unique[(filtered_df_unique['Research Group'] == 'CN') | (filtered_df_unique['Research Group'] == 'MCI') | (filtered_df_unique['Research Group'] == 'LMCI')]
```

```python
# Filter for only MRI scans first
filtered_mri_only = filtered_df_both_subjects[filtered_df_both_subjects['Modality'] == 'MRI']

# Now filter based on matches in 'Subject ID' and 'Age' from filtered_df_unique
# We use isin with a DataFrame built from the relevant columns in filtered_df_unique
matches = filtered_mri_only[filtered_mri_only.set_index(['Subject ID', 'Age']).index.isin(filtered_df_unique.set_index(['Subject ID', 'Age']).index)]

print(matches)
```

```python

# Step 1: Count frequency of each description
frequency = matches['Description'].value_counts().rename('freq')
matches = matches.merge(frequency.to_frame(), left_on='Description', right_index=True)
matches.sort_values(by=['freq', 'Description'], ascending=[False, True], inplace=True)
matches.drop(columns='freq', inplace=True)  # Drop the frequency column if not needed

# Step 2: Prioritize 'Processed' Type
matches['Type_sort'] = matches['Type'].apply(lambda x: 0 if x == 'Processed' else 1)
matches.sort_values(by=['Subject ID', 'Type_sort'], inplace=True)
matches.drop(columns='Type_sort', inplace=True)  # Drop the helper column after sorting

# Step 3: Drop duplicates to keep only the first occurrence per Subject ID
matches_unique = matches.drop_duplicates(subset='Subject ID', keep='first')

print(matches_unique)
```

```python
final_df = pd.concat([filtered_df_unique, matches_unique], ignore_index=True)

print(final_df)
```

```python
# Convert the 'Image ID' column to a list of strings
image_data_ids_list = final_df['Image ID'].tolist()

# Convert each integer ID to a string and then join the list items into a single string, separated by commas
image_data_ids_string = ','.join(str(id) for id in image_data_ids_list)

# image_data_ids_string now contains the IDs separated by commas, as a single string
print(image_data_ids_string)
```
